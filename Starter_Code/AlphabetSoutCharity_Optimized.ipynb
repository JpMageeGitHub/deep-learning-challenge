```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd 
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
application_df = application_df.drop(columns=["EIN", "NAME"])
application_df.head()
```

```python
application_types_to_replace = ["T25", "T14", "T29", "T15", "T17", "T9", "T13", "T9", "T12", "T2"]
# Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
classifications_to_replace = ["C7000", "C1700", "C4000", "C1270", "C2700", "C2800", "C7100", "C1230", "C7200", "C1240", "C8000", "C2300", "C1500", "C7120", "C1800", "C1250", "C1280", "C1238", "C1235", "C1260", "C2710", "C1245", "C1237", "C3200", "C0", "C1257", "C1267", "C1234", "C4120", "C7210", "C1720", "C1256", "C1265", "C2380", "C2570", "C1236", "C1900", "C1246", "C2600", "C2561", "C4500", "C4200", "C5200", "C1728", "C1283", "C6100", "C1820", "C3700", "C2566", "C2577", "C2170", "C2190", "C1239", "C2150", "C1727", "C8210", "C6100", "C2190", "C2150", "C1239", "C1727", "C8210", "C6100", "C2190", "C2150", "C1239", "C1727", "C8210", "C6100", "C2190", "C2150", "C1239", "C1727", "C8210", "C6100", "C2190", "C2150", "C1239", "C1727", "C8210", "C6100", "C2190", "C2150", "C1239", "C1727", "C8210", "C6100", "C2190", "C2150", "C1239", "C1727", "C8210", 
                              "C5000", "C1300", "C1400", "C6000", "C8200", "C1278", "C4100", "C2400", "C1600", "C1732", "C1570", "C2500", "C1580", "C1370", "C1248"]
# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")
    
# Check to make sure binning was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
application_df = pd.get_dummies(application_df)
```

```python
features = application_df.drop(columns=["IS_SUCCESSFUL"])
target = application_df["IS_SUCCESSFUL"]
# Split the preprocessed data into a training and testing dataset
#  YOUR CODE GOES HERE
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=78)
```

```python
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
nn = tf.keras.models.Sequential()

# First hidden layer
#  YOUR CODE GOES HERE
layer1 = tf.keras.layers.Dense(units=100, activation="relu", input_dim= 43)

# Second hidden layer
#  YOUR CODE GOES HERE
layer2 = tf.keras.layers.Dense(units=50, activation="relu")

#Additional hidden layer for optimization
layer3 = tf.keras.layers.Dense(units=25, activation="relu")



# Output layer
#  YOUR CODE GOES HERE
output_layer = tf.keras.layers.Dense(units=1, activation="sigmoid")

nn.add(layer1)
nn.add(layer2)
nn.add(layer3)
nn.add(output_layer)


# Check the structure of the model
nn.summary()
```

```python
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
nn_fit = nn.fit(X_train_scaled, y_train, epochs=100)
```

```python
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
